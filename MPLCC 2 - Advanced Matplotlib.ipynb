```python
fig = plt.figure() # create new matplotlib "Figure" object
# create four subplots (2 rows, 2 columns) and select first subplot
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2) # create and select 2nd subplot
ax3 = fig.add_subplot(2, 2, 3) # create and select 3rd 

# Plot on last subplot used. In this case subplot 3
#'k--' means use color 'k' (black) with dashed lines
plt.plot(np.random.randn(50).cumsum(), 'k--') 
# plot a histogram using first subplot ('ax1')
ax1.hist(np.random.randn(100), bins = 20, color = 'k', alpha = 0.3)
# plot a scatter plot using second subplot('ax2')
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
# Create a NumPy array of subplots using 'plt.subplots()'
fig, axes = plt.subplots(2, 3)
```

<center>Subplot options</center>



```python
# adjust spacing of subplots to 0, share 'x' and 'y' axis amongst subplots
# and address subplots as an array '[x, y]'
# adjust spacing example:
# plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
#                     wspace=None, hspace=None)
fig, axes = plt.subplots(2, 2, sharex = True, sharey = True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace = 0, hspace = 0)
```

```python
# Colors, Markers, and Styles
# refer to docstring by running 'plt.plot?'
# You can specify colors using hex with 'color=' argument (color='#CECECE')
# create a black line plot with dashed lines and circle marker
plt.plot(randn(30).cumsum(), 'ko--')
```

```python
# Draw two line graphs using different styles and place a legend 
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label = 'Default')
plt.plot(data, 'k-', drawstyle = 'steps-post', label = 'steps-post')
plt.legend(loc = 'best') # specify legend location with '.legend()'
```

```python
# Managing a plot's range, tick location and Labels
fig = plt.figure()  # Plot a random walk
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
# Specify where to place ticks within the data range
fig = plt.figure()  # Plot a random walk
ax = fig.add_subplot(1, 1, 1)
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
# Give a label for each tick, use small text and rotate label by 30 degrees
labels = ax.set_xticklabels(['one', 'two', 'three', 'fout', 'five'], 
                             rotation = 30, fontsize = 'small')
ax.set_title('My first matplotlib plot') # Create title
ax.set_xlabel('Stages') # Create x-axis label
ax.plot(np.random.randn(1000).cumsum())
# Can also use dict to set plot properties
props = {'title': 'My first matplotlib plot', 'xlabel': 'Stages'}
ax.set(**props)
```

```python
# Inserting a legend
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label = 'one') # first of 3 line plots
ax.plot(randn(1000).cumsum(), 'k--', label = 'two')
ax.plot(randn(1000).cumsum(), 'k.', label = 'three')
# Create a legend and put it in the 'best' location
ax.legend(loc = 'best') # Can also use 'plt.legend()'
# Note: use 'ax.legend?' for more info on how where to place a legend
```

```python
# Adding annotations
# Example of adding annotations with customized styling
# ax.text(x, y, 'Hello World!', family = 'monospace', fontsize = 10)
fig = plt.figure()  # New plot
ax = fig.add_subplot(1, 1, 1) # Set start of plot location
data = pd.read_csv('examples/spx.csv', index_col = 0, parse_dates = True)
spx = data['SPX']  # load column 'SPX'
spx.plot(ax = ax, style = 'k-')  # Create line chart
crisis_data = [                  # Annotate at specific dates
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Failes'),
    (datetime(2008, 9, 15), 'Leahman Bankruptcy')]

for date, label in crisis_data:  # Add annotations with arrow pointer
    ax.annotate(label, xy = (date, spx.asof(date) + 75),
                xytext = (date, spx.asof(date) + 225),
                arrowprops = dict(facecolor = 'black', headwidth = 4,
                                  width = 2, headlength = 4),
                horizontalalignment = 'left',
                verticalalignment = 'top')
# Set start and end boundaries
ax.set_xlim(['1/1/2007', '1/1/2011']) # set x axis range to desired dates
ax.set_ylim([600, 1800])              # set y axis range to 600 - 1800
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
# Drawing shapes
# While some shapes are available in '.pyplot', a full set is in '.patches'
fig = plt.figure()  # create a new plot
ax = fig.add_subplot(1, 1, 1) # start at origin
# define shapes, a retangle, circle and polygon
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color = 'k', alpha = 0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color = 'b', alpha = 0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], 
                   color = 'g', alpha = 0.5)

ax.add_patch(rect)  # draw shapes
ax.add_patch(circ)
ax.add_patch(pgon)
```
