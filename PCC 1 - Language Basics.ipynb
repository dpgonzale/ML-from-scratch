```python
# Assignments or Binding (references)
a = [1, 2, 3]   # 'a' = a list containing [1, 2, 3]
b = a           # 'b' is a reference to 'a'
c = list(a)
print(b)        # 'c' is a true copy of 'a' 

a.append(4)     # add "4" to a list
print(b)        # print 'b' shows addition of "4" to list
print(c)        # 'c' is not modified since it is a copy of 'a'
```

```python
# You can check if to references refer to the same object with 'is'
print(a is b)
print(a is c)
print(b is not c)
print()

# Note: you check equality use '=='
print(a == c)
a = [1, 2, 3]
print(a == c)
print()

# You can also check objects against None (null) using 'is'
print(a is None)
a = None
print(a is None)
```

```python
# Strings can be delimited by ' ' or " "
a = 'This is a string'
b = "and this is another string"
print(a, b)

# Multiline strings with line breaks use triple quotes: ''' or """
c = """
This is a multiple line string 
with an embedded line break
"""
print(c)

# Strings are immutable (not changeable)
try:
    print(a[10])  # the 10th character of string 'a' is x
    a[10] = 'f'   # try changing the 10th character to an 'f'
except TypeError:
    print('Strings cannot be changed', '\n')
    
# convert python objects to a string using str()
d = 5.6  # a float
s = str(d) # convert to a string
print(s, '\n')

# Strings are sequences and can be sliced. Slices will be addressed later
print(a)
print(a[:3], '\n')

# Use an escape character '\' to use special sharacters such as newline \n
s = '12\\34'
print(s)
# optionaly you can precede a string with 'r' to print in raw format
t = r'12\34'
print(t)
```

```python
# String operations
val = 'a,b, guido'
print(val.split(','))  # split strs based on a delimiter ',' in this case
pieces = [x.strip() for x in val.split(',')] # strip() removes white space
print(pieces)

# concatenate strings using '::' as a delimiter
first, second, third = pieces               # unpack list into variables
print(first + '::' + second + '::' + third) # not pythonic
print('::'.join(pieces), '\n')              # '.join()' is pythonic

print('guido' in val)  # use 'in' to locate substrings
print(val.find(':'))   # Using '.find()'
print(val.index(','))  # Using '.index()'
# Note: Unlike .find() .index fails if string is not found
try:
    print(val.index(':'), '\n')
except ValueError:
    print('-1', '\n')

print(val.count(',')) # '.count()' returns # of occurances of substring
print(val.replace(',', '::')) # '.replace()' substitues occurance of substr
print(val.replace(',', ''))
```

```python
# Object references have no type
a = 5
print(type(a))   # a is an int

a = 'foo'
print(type(a))   # a is now a string

# However Objects are strongly typed
print('5' + 5) 
```

```python
# Objects are not automatically casted to another type
# except in some obvious cirsumstances
a = 4.5
b = 2
print('a is {0}, b is {1}'.format(type(a), type(b)))
c = a / b
print(c)  # b was automatically casted to a float
```

```python
# Conditionaly check an object's type using isinstance()
a = 5
b = 4.5
print(isinstance(a, int))          # check a is an int
print(isinstance(b, (int, float))) # check if b is either an int or a float
```

```python
# Duck typing
# Test is object is of a type by usage and catching the exception
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError: # not an iterable
        return False
```

```python
print(isiterable('a string'))
print(isiterable('[1, 2, 3]'))
print(isiterable(5))
```

```python
# Example use: Check parameter types and convert only if needed
if not isinstance(x, list) and isiterable(x):
    x = list(x)
```

```python
# String Formatting using templates: 
# https://docs.python.org/3.6/library/string.html
template = '{0:.2f} {1:s} are worth US${2:d}'
# {0:.2f} = 1st argument is a floating point number with two decimal places
# {1:s} = 2nd argument is formatted as a string
# {2:d} = 3rd argument is formatted as an integer
# Example:
print(template.format(4.5560, 'Argentine Pesos', 1), '\n')

# Python 3.6 introduces f-string which we will use most often
first_name = 'Han'
last_name = 'Solo'
# old ways are to use: Note string.format is related to templates
full_name1 = first_name + " " + last_name            # string addition
full_name2 = "{0} {1}".format(first_name, last_name) # string.format

# new way: f-string
full_name3 = f"{first_name} {last_name}"
print(full_name1)
print(full_name2)
print(full_name3)
```

```python
# Dates and Times
from datetime import datetime, date, time

dt = datetime(2011, 10, 29, 20, 30, 21)
print(dt.day)
print(dt.minute)
print(dt.date())
print(dt.time())
print(dt.strftime('%m/%d/%Y %H:%M'))           # string format datetime
print(datetime.strptime('20091031', '%Y%m%d')) # parse string into datetime
print(dt.replace(minute=0, second=0))          # zero out minute and seconds of dt
```

```python
# difference between two datetime objects is a timedelta type object
dt2 = datetime(2011, 11, 15, 22, 30)
delta = dt2 - dt
print(type(delta))
print(delta)
delta
```

```python
# adding a timedelta to a datetime creates a new datetime object
print(dt)
print(dt + delta)
```

```python
# for, while continue, break

# For: for value in collection:
#          do something with value
sequence = [1, 2, None, 4, None, 5]
total = 0

for value in sequence:
    if value is None:
        continue       # Skip rest of if block
    total += value
print(total, "\n")     # 12

sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0

for value in sequence:
    if value == 5:
        break           # Break out of for loop when sequence[i] = 5
    total_until_5 += value
print(total_until_5, "\n")  # 13

for i in range(4):
    for j in range(4):  # Range function see below
        if j > i:
            break       # Break out of inner for loop only
        print((i, j))
print()
        
# While: while condition:
#            do something until condition == False
x = 256
total = 0

while x > 0:
    if total > 500:     # loop until x is 0 or less than 0
        break        # or break out of while when total is greater than 500
    total += x
    x = x // 2          # '//' = integer division (i.e. no remainder)    
print(total, x)         # 504, 4  Broke out of loop at total > 500
```
